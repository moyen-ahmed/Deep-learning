```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[2])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

# Create a Sequential model
# Sequential means the layers are added one after another in a linear stack
model = Sequential()

# Flatten Layer
# Input shape is (28, 28) because the image size is 28x28 pixels (like MNIST dataset)
# Flatten converts the 2D image into a 1D vector
# 28 x 28 = 784 values
# Dense layers require 1D input, so this layer prepares the image for the neural network
model.add(Flatten(input_shape=(28,28)))

# First Hidden Layer
# Dense means fully connected layer (every neuron connects to every neuron in previous layer)
# 128 = number of neurons in this layer
# activation='relu' applies the ReLU activation function
# ReLU formula: f(x) = max(0,x)
# It helps the network learn complex patterns and speeds up training
model.add(Dense(128,activation='relu'))

# Second Hidden Layer
# Another Dense layer with 32 neurons
# This layer further processes the features learned from the previous layer
# ReLU is used again because it works well for deep learning models
model.add(Dense(32,activation='relu'))

# Output Layer
# 10 neurons because we have 10 possible classes (digits 0–9)
# Softmax activation converts outputs into probabilities
# The output values will sum to 1
# The class with the highest probability will be the predicted digit
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
