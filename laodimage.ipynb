```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r"C:\Users\estya\OneDrive\Desktop\sunf.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ✅ This shows the pixel array
plt.imshow(img)
np.set_printoptions(threshold=100, edgeitems=3)  # optional formatting
print(img)

# ✅ Optional: print info about the image
print("Shape:", img.shape)       # (height, width, channels)
print("Type:", type(img))        # numpy.ndarray
print("Data type:", img.dtype)   # usually uint8
```
