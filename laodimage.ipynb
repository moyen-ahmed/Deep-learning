# Color Space Conversion

## What is Color Space?

A color space defines the range of colors that can be represented and how they are interpreted. It defines how the color information in the image is structured mathematically.

### Common Color Spaces

*   **RGB**: Represents colors as combinations of red, green, and blue.
*   **CMYK**: Represents colors as combinations of cyan, magenta, yellow, and black (used in printing).
*   **HSV (Hue, Saturation, Value)**: Represents colors in terms of hue (color), saturation (intensity of color), and value (brightness).
*   **LAB**: Represents colors in terms of lightness and color-opponent dimensions.
*   **YUV/YCrCb**: Used in video compression, represents colors as luminance (Y) and chrominance (UV or CrCb).

> *Example*: An image in the HSV color space represents the image using hue, saturation, and value instead of the red, green, and blue channels of the RGB color space.

### Why do we need this in computer vision?

*   Visualization, Different Color Representations
*   Segmentation, Feature Extraction
*   Object Detection, Image Analysis
*   Improved Performance in different algorithms and models

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Load an image (use raw string for path)
image_path = r"C:\Users\estya\OneDrive\Desktop\sunf.jpg"
image = cv2.imread(image_path)

# Convert BGR to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert BGR to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Display the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')

plt.subplot(1, 3, 3)
plt.imshow(hsv_image)
plt.axis('off')
plt.title('HSV Image')




plt.tight_layout()
plt.show()
```
