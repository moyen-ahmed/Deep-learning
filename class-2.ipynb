# Image Thresholding

## What is Image Thresholding?

Image thresholding is a technique in image processing that converts a grayscale image into a binary image. This is done by selecting a threshold value (or multiple threshold values) and classifying the pixels in the image based on their intensity values relative to this threshold.

## Applications of Image Thresholding in Computer Vision

1. **Object Detection**
2. **Image Segmentation**
3. **Character Recognition**
4. **Medical Imaging**
5. **Face Detection**
6. **Motion Detection**
7. **Quality Control**

```python
!pip install opencv-python matplotlib

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Load an image in grayscale
image_path = r"C:\Users\estya\OneDrive\Desktop\orange.jpg"
img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)  # pass flag here

# Apply different thresholding techniques
ret , th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
ret , th2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
ret , th3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
ret , th4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)
ret , th5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)
th6 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
th7 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.figure(figsize=(10,7))
titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV', 'ADAPTIVE MEAN', 'ADAPTIVE GAUSSIAN']
images = [img, th1, th2, th3, th4, th5, th6, th7]
for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
