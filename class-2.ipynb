# Image Thresholding

## What is Image Thresholding?

Image thresholding is a technique in image processing that converts a grayscale image into a binary image. This is done by selecting a threshold value (or multiple threshold values) and classifying the pixels in the image based on their intensity values relative to this threshold.

## Applications of Image Thresholding in Computer Vision

1. **Object Detection**
2. **Image Segmentation**
3. **Character Recognition**
4. **Medical Imaging**
5. **Face Detection**
6. **Motion Detection**
7. **Quality Control**

```python
!pip install opencv-python matplotlib

import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Load an image in grayscale
image_path = r"C:\Users\estya\OneDrive\Desktop\orange.jpg"
img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)  # pass flag here

# Apply different thresholding techniques
ret, thresh1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
ret, thresh2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
ret, thresh3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)      # >threshold = threshold; <= unchanged
ret, thresh4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)     # >threshold unchanged; <= 0
ret, thresh5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV) # >threshold = 0; <= unchanged

# Titles and images
titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

# Display all images
plt.figure(figsize=(15, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=255)
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])

plt.tight_layout()
plt.show()
```
